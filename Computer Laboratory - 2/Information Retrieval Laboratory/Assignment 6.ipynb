Q. Build the web crawler to pull IMDB information and analyse the
ratings.

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.imdb.com/chart/top/"
```

```python
HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
```

```python
page=requests.get(url,headers=HEADERS)
page
```

```python
soup=BeautifulSoup(page.content,"html.parser")
```

```python
scrapped_movies=soup.find_all('h3',class_='ipc-title__text')
scrapped_movies
```

```python
movies=[]
for movie in scrapped_movies:
    movie=movie.get_text().replace('\n',"")
    movie=movie.strip(" ")
    movies.append(movie)
movies
```

```python
scrapped_ratings=soup.find_all('span',class_="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating")
scrapped_ratings
```

```python
ratings=[]
for rating in scrapped_ratings:
    rating=rating.get_text().replace('\n','')
    rating = rating.strip(" ")
    ratings.append(rating[:3])
ratings
```

```python
import pandas as pd
data=pd.DataFrame()
movies = movies[:250]
data['Movie Names']= movies
data['Ratings']=ratings
data.head()
```

```python
def crawl_web(seed_url,max_pages):
    pages_visited=0
    url_queue=[seed_url]
    
    while pages_visited < max_pages and url_queue:
        url = url_queue.pop(0)
        
        try:
            response = requests.get(url)
            if response.status_code ==200:
                soup=BeautifulSoup(response.text,'html.parser')
                
                print(soup.title.string)
                
                links=[a['href']for a in soup.find_all('a',href=True)]
                for link in links:
                    url_queue.append(link)
                    
                pages_visited += 1
        except Exveption as e:
            print(f"Error:{e}")
#         e usage:
crawl_web("https://google.com",max_pages=6)
                
            
```
